```python
#importing Necessary Libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#reading data file
data = pd.read_csv("omicron.csv")
data.head()
```

```python
#converting Data to a Pandas Dataframe
df = pd.DataFrame(data)
```

```python
df.info
```

```python
df.describe()
```

```python
# checking how many null data are present
df.isnull().sum()
```

```python
# extracting United states Data only
usa = df [df["Country"] == 'United States']
```

```python
usa
```

```python
# Plotting Line graph using seaborn and matplotlib
fig, axes = plt.subplots(figsize = (12,8))
plt.title("Omicron in USA: Aug-Dec")
sns.lineplot(x = "Date", y = "Percentage of Omicron", data = usa, ax = axes)
```

```python
# extracting data of Six Countries
Countries = ["United States", "Germany", "Japan", "Russia", "Canada","India"]
six = df[(df['Country'] == "United States") | (df['Country'] == "Germany") |(df['Country'] == "Japan") | (df['Country'] == "Canada") |(df['Country'] == "India")]
```

```python
six
```

```python
#plottting line graph of extrated countries
fig, axes = plt. subplots (figsize = (14,8)) 
plt.title ("Omicron distribution in six countries")
sns.lineplot (x = "Date", y = "Percentage of Omicron", hue = "Country", data = six, ax = axes)
```
